```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_feather('../Data/dicom_df.feather')
df['ContrastFlowRate']
```

```python
df1 = pd.read_feather('../1_Data_Extraction/Data/Metadata_Only/dicom_df_0.feather')
df2 = pd.read_feather('../1_Data_Extraction/Data/Metadata_Only/dicom_df_1.feather')
df3 = pd.read_feather('../1_Data_Extraction/Data/Metadata_Only/dicom_df_2.feather')
df4 = pd.read_feather('../1_Data_Extraction/Data/Metadata_Only/dicom_df_3.feather')
df5 = pd.read_feather('../1_Data_Extraction/Data/Metadata_Only/dicom_df_4.feather')
df6 = pd.read_feather('../1_Data_Extraction/Data/Metadata_Only/dicom_df_5.feather')
df7 = pd.read_feather('../1_Data_Extraction/Data/Metadata_Only/dicom_df_6.feather')
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
```

```python
df.head()
```

```python
df1.info()
```

```python
df_small = df[["PatientId","PatientWeight","PatientSize","PatientAge","PatientSex", "BodyPartExamined", "ProcedureCodeSequence.CodeMeaning","Rows","Columns","SliceCount","PixelSpacing","SliceThickness"]]
df_small
```

```python
df_cleaned = df_small.copy()

df_cleaned.loc[:, "PatientAge"] = df_cleaned["PatientAge"].fillna('-1')
df_cleaned.loc[:, "PatientAge"] = df_cleaned["PatientAge"].str.replace('Y', '').astype(int)
df_cleaned.loc[:, "BodyPart"]   = df_cleaned["ProcedureCodeSequence.CodeMeaning"].str.split(".", expand=True)[2]
```

```python
df_cleaned.describe()
```

```python
# Distribution of Male/female
df_cleaned['PatientSex'].value_counts()
```

```python
# Distribution of Age
df_cleaned['PatientAge'].value_counts()
```

```python
# Get all lines where BodyPart is Ganzkörper and SliceCount is greater than 600
df_cleaned[(df_cleaned['BodyPart'] == 'Ganzkörper') & (df_cleaned['SliceCount'] > 600)]
```

```python
# Get all lines where BodyPart is Herz
df_cleaned[df_cleaned['BodyPart'] == 'Herz']
```

```python
# Distribution of Sex
plt.figure(figsize=(15, 6))
sns.countplot(x='PatientSex', data=df)
plt.title('Distribution of Patient Sex')
plt.xlabel('Sex')
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Patient Age
plt.figure(figsize=(15, 6))
sns.histplot(df_cleaned['PatientAge'], kde=True)
plt.title('Distribution of Patient Age')
plt.xlabel('Age (years)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Patient Weight
plt.figure(figsize=(15, 6))
sns.histplot(df_cleaned['PatientWeight'], kde=True)
plt.title('Distribution of Patient Weight')
plt.xlabel('Weight (kg)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
```

```python
# Distribution of Patient Size
plt.figure(figsize=(15, 6))
sns.histplot(df_cleaned['PatientSize'], kde=True)
plt.title('Distribution of Patient Size')
plt.xlabel('Size (m)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
```

```python
# SliceCount by BodyPartExamined
plt.figure(figsize=(18, 8))
sns.boxplot(x='BodyPartExamined', y='SliceCount', data=df_cleaned)
plt.title('Slice Count by Body Part Examined')
plt.xlabel('Body Part Examined')
plt.xticks(rotation=90)
plt.ylabel('Slice Count')
plt.show()
```

```python
# BodyPart distribution histogram descending frequency
plt.figure(figsize=(15, 6))
sns.countplot(y='BodyPart', data=df_cleaned, order = df_cleaned['BodyPart'].value_counts().index)
plt.title('Distribution of Body Part')
plt.xlabel('Frequency')
plt.ylabel('Body Part')
plt.show()
```

```python
# SliceCount by BodyPart
plt.figure(figsize=(18, 8))
sns.boxplot(x='BodyPart', y='SliceCount', data=df_cleaned)
plt.title('Slice Count by Body Part')
plt.xlabel('Body Part')
plt.xticks(rotation=90)
plt.ylabel('Slice Count')
plt.show()
```

```python
# SliceCount distribution
plt.figure(figsize=(15, 6))
sns.histplot(df_cleaned['SliceCount'], kde=True)
plt.title('Distribution of Slice Count')
plt.xlabel('Slice Count')
plt.xticks(rotation=90)
plt.ylabel('Frequency')

plt.show()
```

```python
# SliceThickness distribution
plt.figure(figsize=(15, 6))
sns.histplot(df_cleaned['SliceThickness'], kde=True)
plt.title('Distribution of Slice Thickness')
plt.xlabel('Slice Thickness (mm)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
```

```python
# get lists of all x and y pixel spacings with array splicing
x_pixel_spacing = df_cleaned['PixelSpacing'].apply(lambda x: x[0])
y_pixel_spacing = df_cleaned['PixelSpacing'].apply(lambda x: x[1])
```

```python
# PixelSpacing in x direction distribution
plt.figure(figsize=(15, 6))
sns.histplot(x_pixel_spacing, kde=True)
plt.title('Distribution of Pixel Spacing in x direction')
plt.xlabel('Pixel Spacing (mm)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
```

```python
# PixelSpacing in y direction distribution
plt.figure(figsize=(15, 6))
sns.histplot(y_pixel_spacing, kde=True)
plt.title('Distribution of Pixel Spacing in x/y direction')
plt.xlabel('Pixel Spacing (mm)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
```
