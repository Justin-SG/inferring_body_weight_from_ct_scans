```python
import pandas as pd
%matplotlib inline
```

```python
df1 = pd.read_feather('../1_Data_Extraction/Data/dicom_df_0.feather')
df2 = pd.read_feather('../1_Data_Extraction/Data/dicom_df_1.feather')
df3 = pd.read_feather('../1_Data_Extraction/Data/dicom_df_2.feather')
df4 = pd.read_feather('../1_Data_Extraction/Data/dicom_df_3.feather')
df5 = pd.read_feather('../1_Data_Extraction/Data/dicom_df_4.feather')
df6 = pd.read_feather('../1_Data_Extraction/Data/dicom_df_5.feather')
df7 = pd.read_feather('../1_Data_Extraction/Data/dicom_df_6.feather')
```

```python
df = pd.concat([df1, df2, df3, df4, df5, df6, df7])
```

```python
df.head()
```

```python
df.drop(columns=['PixelArrayFlat'], inplace=True)
```

```python
df1.info()
```

```python
df_small = df[["PatientId","PatientWeight","PatientSize","PatientAge","PatientSex", "BodyPartExamined", "ProcedureCodeSequence.CodeMeaning","Rows","Columns","SliceCount","PixelSpacing","SliceThickness"]]
df_small
```

```python
df_small["PatientAge"].fillna("-1", inplace=True)
```

```python
df_small["PatientAge"] = df_small["PatientAge"].str.replace('Y', '').astype(int)
```

```python
df_small["BodyPart"] = df_small["ProcedureCodeSequence.CodeMeaning"].str.split(".", expand=True)[2]
```

```python
import matplotlib.pyplot as plt

scan = df1.iloc[2]
array = scan['PixelArrayFlat'].reshape((scan['SliceCount'], scan['Rows'], scan['Columns']))
print("Array shape:", array.shape)
print("PixelSpacing:", scan['PixelSpacing'], "SliceThickness:", scan['SliceThickness'])
# pixel aspects, assuming all slices are the same
pixel_spacing = scan.PixelSpacing
slice_thickness = scan.SliceThickness
ax_aspect = pixel_spacing[1] / pixel_spacing[0]
sag_aspect = pixel_spacing[1] / slice_thickness
cor_aspect = slice_thickness / pixel_spacing[0]
```

```python
from matplotlib import animation

scan3d = array
frames = scan['SliceCount']

fig, ax = plt.subplots()
im = plt.imshow(scan3d[0,:,:], cmap=plt.cm.bone, aspect=ax_aspect)
ax.set_title("Axial")

def update(frame):
    im.set_array(scan3d[frame, :, :])
    return im


anim = animation.FuncAnimation(fig=fig, func=update, frames=frames, interval=100)
plt.show()
```

```python
scan3d = array
frames = scan['Rows']

fig, ax = plt.subplots()
im = plt.imshow(scan3d[:,0,:], cmap=plt.cm.bone, aspect=cor_aspect)
ax.set_title("Frontal")

def update(frame):
    im.set_array(scan3d[:, frame, :])
    return im

anim = animation.FuncAnimation(fig=fig, func=update, frames=frames, interval=100)
plt.show()
```

```python
scan3d = array
frames = scan['Columns']

fig, ax = plt.subplots()
im = plt.imshow(scan3d[:,:,0].T, cmap=plt.cm.bone, aspect=sag_aspect)
ax.set_title("Saggital")

def update(frame):
    im.set_array(scan3d[:, :, frame])
    return im

anim = animation.FuncAnimation(fig=fig, func=update, frames=frames, interval=100)
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
 
# Distribution of Patient Age
plt.figure(figsize=(15, 6))
sns.histplot(df_small['PatientAge'], kde=True)
plt.title('Distribution of Patient Age')
plt.xlabel('Age (years)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
 
# Distribution of Patient Weight
plt.figure(figsize=(15, 6))
sns.histplot(df_small['PatientWeight'], kde=True)
plt.title('Distribution of Patient Weight')
plt.xlabel('Weight (kg)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
 
# Distribution of Patient Size
plt.figure(figsize=(15, 6))
sns.histplot(df_small['PatientSize'], kde=True)
plt.title('Distribution of Patient Size')
plt.xlabel('Size (m)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
 
# SliceCount by BodyPartExamined
plt.figure(figsize=(18, 8))
sns.boxplot(x='BodyPartExamined', y='SliceCount', data=df_small)
plt.title('Slice Count by Body Part Examined')
plt.xlabel('Body Part Examined')
plt.xticks(rotation=90)
plt.ylabel('Slice Count')
plt.show()

# BodyPart distribution
plt.figure(figsize=(15, 6))
sns.countplot(x='BodyPart', data=df_small)
plt.title('Distribution of Body Part')
plt.xlabel('Body Part')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()

# SliceCount by BodyPart
plt.figure(figsize=(18, 8))
sns.boxplot(x='BodyPart', y='SliceCount', data=df_small)
plt.title('Slice Count by Body Part')
plt.xlabel('Body Part')
plt.xticks(rotation=90)
plt.ylabel('Slice Count')
plt.show()
 
# SliceThickness distribution
plt.figure(figsize=(15, 6))
sns.histplot(df_small['SliceThickness'], kde=True)
plt.title('Distribution of Slice Thickness')
plt.xlabel('Slice Thickness (mm)')
plt.xticks(rotation=90)
plt.ylabel('Frequency')
plt.show()
```
