```python
from pathlib import Path
import os
import pandas as pd
```

```python
PATH_TO_DIR = Path(os.getcwd()).resolve().parent.parent

PATH_TO_FLOS_EVALUATION = PATH_TO_DIR / "5_Evaluation" / "Flo" / "model_predictions.feather"
```

```python
PATH_TO_FLOS_EVALUATION
```

```python
# Open feather file
flos_evaluation = pd.read_feather(PATH_TO_FLOS_EVALUATION)
flos_evaluation
```

```python
flos_evaluation.iloc[0]
```

```python
df_unique = flos_evaluation.drop_duplicates(subset=['set_type', 'patient_id', 'pixel_array_file'])
```

```python
df_unique
```

```python
PATH_TO_DATA_DIR = Path(os.getcwd()).resolve().parent.parent / "Data"
PATH_TO_CLEANED_DICOM_DF = PATH_TO_DATA_DIR / "cleaned_dicom_df.feather"

cleaned_dicom_df = pd.read_feather(PATH_TO_CLEANED_DICOM_DF)

cleaned_dicom_df = cleaned_dicom_df[cleaned_dicom_df["BodyPart"] == "Stamm"]
cleaned_dicom_df
```

```python
PATH_TO_DATA_DIR = Path(os.getcwd()).resolve().parent.parent / "Data"
PATH_TO_CLEANED_DICOM_DF = PATH_TO_DATA_DIR / "cleaned_dicom_df_ORIGINAL_DO_NOT_DELETE.feather"

cleaned_dicom_df = pd.read_feather(PATH_TO_CLEANED_DICOM_DF)

cleaned_dicom_df = cleaned_dicom_df[cleaned_dicom_df["BodyPart"] == "Stamm"]
cleaned_dicom_df
```

```python
# Print how many unique values are in the column PatientId
print(cleaned_dicom_df["SeriesInstanceUID"].nunique())
```

```python
# join cleaned_dicom_df and df_unique on column PixelArrayFile
joined_df = cleaned_dicom_df.merge(df_unique, left_on="PixelArrayFile", right_on="pixel_array_file")
joined_df
```

```python
# Drop all columns except for the following: SeriesInstanceUID, set_type
joined_df = joined_df[["SeriesInstanceUID", "set_type"]]
joined_df
```

```python
# count unique values in column SeriesInstanceUID
joined_df["SeriesInstanceUID"].nunique()
```

```python
# Save joined_df as feather file
PATH_TO_JOINED_DF = PATH_TO_DATA_DIR / "train_test_split.feather"
joined_df.to_feather(PATH_TO_JOINED_DF, version=2, compression="zstd")
```
