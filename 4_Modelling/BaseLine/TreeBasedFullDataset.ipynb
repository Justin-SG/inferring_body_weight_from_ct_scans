```python
from pathlib import Path
import os
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeRegressor
```

```python
PATH_TO_PARENT_DIR = Path(os.getcwd()).resolve().parent.parent

PATH_TO_DATA_DIR = PATH_TO_PARENT_DIR / "Data"
PATH_TO_CLEANED_DICOM_DF = PATH_TO_DATA_DIR / "cleaned_dicom_df.feather"
PATH_TO_TRAIN_TEST_SPLIT = PATH_TO_DATA_DIR / "train_test_split.feather"
```

```python
# Laod PATH_TO_CLEANED_DICOM_DF feather file
meta_df = pd.read_feather(PATH_TO_CLEANED_DICOM_DF)
train_test_split_df = pd.read_feather(PATH_TO_TRAIN_TEST_SPLIT)
meta_df['PatientSex'] = meta_df['PatientSex'].apply(lambda x: 0 if x == 'M' else 1)

meta_df
```

```python
training_df = train_test_split_df[train_test_split_df["set_type"] == "Train"]
training_df
```

```python
test_df = train_test_split_df[train_test_split_df["set_type"] == "Validation"]
test_df
```

```python
# Get PatientAge	PatientSex	PatientWeight	PatientSize and SeriesInstanceUID
data_df = meta_df[["PatientAge", "PatientSex", "PatientWeight", "PatientSize", "SeriesInstanceUID"]]
data_df
```

```python
training_data_df = pd.merge(training_df, data_df, on="SeriesInstanceUID")
target = training_data_df["PatientWeight"]
# drop set_type and SeriesInstanceUID and PatientWeight
training_data_df = training_data_df.drop(columns=["set_type", "SeriesInstanceUID", "PatientWeight"])
training_data_df
```

```python
# train decision tree model with best parameters
dt = DecisionTreeRegressor(random_state=42)
dt.fit(training_data_df, target)
```

```python
testing_data_df = pd.merge(test_df, data_df, on="SeriesInstanceUID")
target = testing_data_df["PatientWeight"]
# drop set_type and SeriesInstanceUID and PatientWeight
testing_data_df = testing_data_df.drop(columns=["set_type", "SeriesInstanceUID", "PatientWeight"])
testing_data_df
```

```python
predictions = dt.predict(testing_data_df)

# Create df with columns Model="Baseline",Prediction, Actual ,Deviation
results_df = pd.DataFrame()
results_df["Model"] = ["Baseline"] * len(predictions)
results_df["Prediction"] = predictions
results_df["Actual"] = target.values
results_df["Deviation"] = results_df["Prediction"] - results_df["Actual"]
results_df.to_csv("baseline_results.csv", index=False)
```
