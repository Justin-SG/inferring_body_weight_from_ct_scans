```python
# Import necessary libraries
import os
import pydicom as dicom
from zipfile import ZipFile
from matplotlib import pyplot as plt

# Set the scan identifier and paths
scan = 'CT_example/1001EE27'

# List DICOM files in the target directory
slices = os.listdir(scan)

# Read the first DICOM file in the series
path = f'{scan}/{slices[0]}'
scan_frame = dicom.dcmread(path, force=True)

# Print DICOM attributes
# Print all attributes
print("All attributes:")
print(scan_frame.dir())
print("=" * 100)

# Print specific attributes
for attribute in scan_frame.dir():
    if(attribute == 'PixelData'):
        continue
    print(attribute, ":", getattr(scan_frame, attribute))
print("=" * 100)

# Display the DICOM image using matplotlib
print("Displaying the DICOM image...")
plt.imshow(scan_frame.pixel_array, cmap=plt.cm.grey)
plt.title(path)
plt.show()
```
