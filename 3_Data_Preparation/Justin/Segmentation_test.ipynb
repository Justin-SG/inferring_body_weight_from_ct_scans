```python
# Call TotalSegmentation with the folder path of each BodyPart

from totalsegmentator.python_api import totalsegmentator
import os
import pandas as pd
import nibabel as nib

output_path = "C:/Users/schoe/Desktop/inferring_body_weight_from_ct_scans/3_Data_Preparation/Justin/TS_Output/"

for body_part, meta_data in df_grouped.iterrows():
    print(body_part, meta_data.FolderPath)
    output_folder = output_path + body_part
    totalsegmentator(meta_data.FolderPath, output_path + body_part, fast=True)
    
    # Read all nii.gz files in the output folder and convert them to a dataframe
    # The dataframe columns are the file names and the values are the sum of the voxel values in the file
    # Save the dataframe as a feather file
    file_names = []
    voxel_sums = []

    for file_name in os.listdir(output_folder):
        if file_name.endswith(".nii.gz"):
            file_path = os.path.join(output_folder, file_name)
            img = nib.load(file_path)
            voxel_sum = img.get_fdata().sum()
            file_names.append(file_name)
            voxel_sums.append(voxel_sum)

    df = pd.DataFrame({"File Name": file_names, "Voxel Sum": voxel_sums})

    # Save the dataframe as a feather file
    df.to_feather(output_folder + "_output_dataframe.feather")
```

```python
BINS = {
    "Air": (-1000, -1000),
    "Fat": (-120, -90),
    "Soft tissue on contrast CT": (100, 300),
    "Bone Cancellous": (300, 400),
    "Bone Cortical": (500, 1900),
    "Subdural hematoma (First hours)": (75, 100),
    "Subdural hematoma (After 3 days)": (65, 85),
    "Subdural hematoma (After 10–14 days)": (35, 40),
    "Other blood Unclotted": (13, 50),
    "Other blood Clotted": (50, 75),
    "Pleural effusion Transudate": (2, 15),
    "Pleural effusion Exudate": (4, 33),
    "Other fluids Chyle": (-30, -30),
    "Water": (0, 0),
    "Urine": (-5, 15),
    "Bile": (-5, 15),
    "CSF": (15, 15),
    "Abscess/Pus": (0, 45),
    "Mucus": (0, 130),
    "Lung Parenchyma": (-700, -600),
    "Kidney": (20, 45),
    "Liver": (54, 66),  # 60 ± 6
    "Lymph nodes": (10, 20),
    "Muscle": (35, 55),
    "Thymus (Children)": (20, 40),
    "Thymus (Adolescents)": (20, 120),
    "White matter": (20, 30),
    "Grey matter": (37, 45),
    "Gallstone Cholesterol stone": (30, 100),
    "Gallstone Bilirubin stone": (90, 120),
    "Foreign body Windowpane glass": (500, 500),
    "Foreign body Aluminum, tarmac, etc.": (2100, 2300),
    "Foreign body Limestone": (2800, 2800),
    "Foreign body Copper": (14000, 14000),
    "Foreign body Silver": (17000, 17000),
    "Foreign body Steel": (20000, 20000),
    "Foreign body Gold/Steel/Brass": (30000, 30000)
}

def bin_hounsfield_values(image_array_1d, hounsfield_scale):
    # Initialize dictionary to store bin counts
    bin_counts = {key: 0 for key in hounsfield_scale.keys()}
    
    # Iterate over each range in the Hounsfield scale
    for label, (lower_bound, upper_bound) in hounsfield_scale.items():
        if lower_bound is None:
            continue  # Skip undefined ranges
        
        # Create a mask for values that fall within the current Hounsfield range
        mask = (image_array_1d - 1000 >= lower_bound - 1) & (image_array_1d - 1000 <= upper_bound + 1)
        
        # Count how many values fall in this range and store the result
        bin_counts[label] = np.sum(mask)

    return bin_counts

result = bin_hounsfield_values(image_array_1d, BINS)

# Output the bin counts
for tissue_type, count in result.items():
    print(f"{tissue_type}: {count}")
```
